IN ORDER TO SCRAPE THE TABLE, WE CAN USE BEAUTIFULSOUP. THE WEBPAGE TITLE IS PRINTED TO CHECK IF THE PAGE HAS BEEN SCRAPED SUCCESSFULLY

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

WE CONVERT THE TABLE TO DATAFRAME FOR CLEANING AND PREPROCESSING

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

NOW, WE PREPROCESS AND CLEAN THE DATA 

```python
# Dropping the rows where Borough is 'Not assigned'
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])

df2
```

FINALLY, WE OBTAIN THE SHAPE OF THE DATAFRAME

```python
df2.shape
```

WE IMPORT THE CSV FILE THAT CONTAINS THE LONGITUDES AND LATITUDES FOR TORONTO POSTCODES

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

WE MERGE THE TWO TABLES IN ORDER TO SHOW THE LONGITUDES AND LATITUDES FOR TORONTO NEIGHBOURHOODS

```python
df3 = pd.merge(df2,lat_lon,on='Postal Code')
df3.head(12)
```

NOW WE WILL CLUSTER THE NEIGHBOURHOODS IN TORONTO. WE WILL FIRSTLY NARROW DOWN THE DATA SO THAT WE ARE WORKING WITH BOROUGHS THAT CONTAIN THE WORD 'TORONTO'

```python
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

NOW, WE WILL VISUALISE THE NEIGHBOURHOODS FROM THE ABOVE DATA USING THE FOLIUM PACKAGE. WE WILL ADD MARKERS TO THE MAP IN ORDER TO SHOW WHERE THE NEIGHBOURHOODS ARE LOCATED

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighborhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

WE WILL RUN K-MEANS TO CLUSTER THE NEIGHBOURHOODS INTO 5 CLUSTERS, CHECKING THAT THE CLUSTER LABELS ARE GENERATED FOR EACH ROW IN THE DATAFRAME

```python
k=5
toronto_clustering = df4.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
```

FINALLY, WE WILL CREATE THE MAP, SET THE COLOUR SCHEME FOR THE CLUSTERS AND ADD MARKERS TO THE MAP

```python
map_clusters = folium.Map(location=[43.651070,-79.347015], zoom_start=11)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighborhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighborhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
